<h1>Public Opinion Results</h1>

<h3> Introduction </h3>
This is an exploratory data science project tracking public opinion on Twitter of three technology companies (Google, Microsoft, and Tesla) over the course of 2021. I used the Python library snscrape to gather 11,000 tweets in English tweeted over the course of 11 months that mention each of the companies and then used a natural language processing sentiment analysis model to numericalize the tweets in terms of how positive or negative their tone was.

<h3> Loading Data </h3>
The final data is stored in a spreadsheet titled final_data.csv. It contains the means and standard deviations of the three companies over an eleven-month period. To load it, the proper libraries need to be imported, and then the CSV files have to be read by the pandas library into three DataFrames, each with a column for each month, and two columns for mean and standard deviation, respectively.

```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
```

```python
results = pd.read_csv('final_data.csv')
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November']
google = pd.DataFrame([months, results['Google Mean'], results['Google Standard Deviation']], 
                      ['Month', 'Mean', 'Standard Deviation']).transpose()
microsoft = pd.DataFrame([months, results['Microsoft Mean'], results['Microsoft Standard Deviation']], 
                         ['Month', 'Mean', 'Standard Deviation']).transpose()
tesla = pd.DataFrame([months, results['Tesla Mean'], results['Tesla Standard Deviation']], 
                     ['Month', 'Mean', 'Standard Deviation']).transpose()
```

To get a more intuitive feel for this data, we can use the matplotlib Python library to create graphs of the data. The first graph is of the mean positivity of each tweet that mentions the company over each month, and the second is the standard deviation of each month's data.

```python
plt.plot(months, google['Mean'], label='Google')
plt.plot(months, microsoft['Mean'], label='Microsoft')
plt.plot(months, tesla['Mean'], label='Tesla')
plt.xlabel('Time (months)')
plt.ylabel('Positivity of public opinion (scale from 0-1)')
plt.title('Twitter Public Opinion of Technology Companies\n in 2021 over Time')
plt.legend()
ax = plt.gca()
ax.xaxis.set_major_locator(matplotlib.ticker.FixedLocator(range(11)))
ax.set_xticklabels(labels=months,rotation=60);
plt.show()
```

```python
plt.plot(months, google['Standard Deviation'], label='Google')
plt.plot(months, microsoft['Standard Deviation'], label='Microsoft')
plt.plot(months, tesla['Standard Deviation'], label='Tesla')
plt.xlabel('Time (months)')
plt.ylabel('Standard deviation of public opinion')
plt.title('Standard Deviation of Twitter Public Opinion of \nTechnology Companies over Time')
plt.legend()
ax = plt.gca()
ax.xaxis.set_major_locator(matplotlib.ticker.FixedLocator(range(11)))
ax.set_xticklabels(labels=months,rotation=60);
plt.show()
```

<h3> Results </h3>
The first interesting part of the data would be that none of the companies every reached 31% positivity, suggesting that the average tweet regarding these companies was not positive. The data would also suggest that the Twitter public has a lower view of Google than Tesla and Microsoft, with the highest value Google reached being 20.5% positivity, lower than the minima of the two other companies, 22.5% and 25.1% for Microsoft and Tesla, respectively. This is interesting to me, given that I was initially expecting Tesla to have the lowest scores, because of the status of its CEO, Elon Musk, known for expressing controversial opinions very openly on social media. Google also consistently had the lowest standard deviation, meaning there were less opinions spread far away from the mean opinion on Google than there were on other companies.
