```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
```

```python
unwanted = ['Sex', 'Municipality', 'Sector', 'Year']
```

```python
def time_plot(x, y, xlabel, ylabel, gender):
    plt.plot(x, y, label=gender)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(f'{wanted} over Time')
```

```python
df = pd.read_csv('Year.csv')
wanted = [item for item in df.columns.values if item not in unwanted]
man_df = df[df['Sex'] == 'Hombre']
woman_df = df[df['Sex'] == 'Mujer']
for wanted in wanted:
    plt.figure()
    time_plot(man_df['Year'], list(man_df[wanted]), 'Year', wanted, 'Male')
    time_plot(woman_df['Year'], list(woman_df[wanted]), 'Year', wanted, 'Female')
    plt.legend()
    plt.show()
```
